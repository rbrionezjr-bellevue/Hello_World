**Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.**

```python
y = dataframe['User_Score'].replace('tbd', np.nan)
# saving the returned pandas series as a variable 
```

```python
y.unique()
# verifying that the string 'tbd' no longer is present
```

```python
dataframe = dataframe.assign(User_Score = y)
# assigning the pandas series 'y' as a new column (overwriting the existing) 'User_Score'
dataframe['User_Score'].unique()
# checking the unique values of 'User_Score' to be sure the string 'tbd' still does not exist
```

```python
dataframe['User_Score'].median()
# using median to calculate the median 'User_Score'
```

```python
z = dataframe['User_Score'].replace(np.nan, '7.5')
# using replace() to repalce NaN values with 7.5 which was the calcualted median and saving to variable 'z'
dataframe = dataframe.assign(User_Score = z)
# assignign the new pandas series saved to z as 'User_Score'
dataframe.head(20)
# Viewing the first 20 rows to check the column 'User_Score'
```
