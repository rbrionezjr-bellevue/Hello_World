## Graph Analysis of Data

```python
# importing required libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
viridis = mpl.colormaps['viridis'].resampled(8)
```

```python
# plotting a histogram of the 'type_1' column in the data frame
plt.hist(df['Type_1'],edgecolor= "black",align='mid', bins=np.arange(19)-0.5)
plt.xticks(rotation = 90)
plt.xlabel("Type of Pokemon")
plt.ylabel("Count")
plt.title("Count of Types of Pokemon")
plt.show()
```

From the histogram of types, we can see that there are certian types of pokem that exist more often than other types. The histogram shows that the water type pokemon is the most abundant in total across all pokemon games.

Without doing a full analysis of the correlation of the two variables, it visually appears that there is a slightly positive correlation between attack power and defense.

```python
types_attack = round(df.groupby(['Type_1'])['Attack'].mean(),3)
types_defense = round(df.groupby(['Type_1'])['Defense'].mean(),3)
types_health = round(df.groupby(['Type_1'])['HP'].mean(),2)
types_attack_df = types_attack.to_frame().reset_index() 
types_defense_df = types_defense.to_frame().reset_index() 
types_health_df = types_health.to_frame().reset_index() 
X = types_health_df['Type_1'].tolist()
```

```python
# importing additional libraries
import seaborn as sns
```

```python
# plotting a box plot to further evaluate the attack power of pokemon types
sns.set_style("whitegrid") 
sns.boxplot(x = 'Type_1', y = 'Attack', data = df) 
plt.xticks(rotation=90)
plt.xlabel('Type of Pokemon')
plt.ylabel('Attack Power')
plt.title('Exploring Attack Power by Pokemon Type')
plt.show()
```

Using a box plot is a good way to find outliers and visually illsutrate the range of the data. We can easily see the min and max attack power of each type of pokemon.

### Conclusion of Milestone 1

After reviewing the graphical analysis, we can see there are some unbalanced attributes of various Pokémon types. When reviewing the histogram of types of Pokémon, we can see that the water type is the most prevalent type of Pokémon available.  When reviewing the scatterplot of attack power by Pokémon type and number we can see there does not appear to be any obvious correlation. Since the number of the Pokémon is an indication of generation (the generation relating to the year of release for the Pokémon game), we could say that the more recently released Pokémon games do not include “stronger” Pokémon, which could be an indication of balance. Thus far there does not appear to be any significant indication of unbalanced gameplay in regard to Pokémon. 

Additional analysis of the overall data set must be done to be sure there is not an imbalance in gameplay regarding Pokémon. We can see there are a few outliers when reviewing the boxplot titled “'Exploring Attack Power by Pokemon Type”. We can see that the ‘Fairy” type Pokémon has two extreme outliers in attack power above the maximum noted attack power value for that type. Additionally, we see outliers above the maximum attack power with the ‘Normal’, ‘Psychic’, and ‘Steel’ type Pokémon. These outliers could indicate an issue with imbalance for these types of Pokémon.
