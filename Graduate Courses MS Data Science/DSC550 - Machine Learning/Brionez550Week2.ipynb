```python
# Using the matplotlib plt.hist function to plot a histogram of the count of "Platforms"
plt.hist(df["Platform"], color="green" ,edgecolor= "black", bins=31)
# label for x axis
plt.xlabel("Platorm")
# label fo y axis
plt.ylabel("Count")
# plot title
plt.title("Histogram of Games by Platform")
# rotation of x tick marks to 90 degrees for easier readability
plt.xticks(rotation=90)
plt.show()
```

```python
# The code below operates the same as the commented code above
plt.hist(df["Year_of_Release"], color = "purple", edgecolor= "black", bins=30)
plt.xlabel("Year of Release")
plt.ylabel("Count")
plt.title("Histogram of Games by Release Year")
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating a new dataframe that only includes games for the "Platform" NES
nes = df[df['Platform'] == "NES"]
```

```python
# Creating a new dataframe that only includes games for the "Platform" SNES
snes = df[df['Platform'] == "SNES"]
```

```python
# Creating a new dataframe that only includes games for the "Platform" N64
n64 = df[df['Platform'] == "N64"]
```

```python
# Using plt.bar to plot the global sales of the top 10 games specific to N64 Platform
plt.bar(n64['Name'].head(10), n64['Global_Sales'].head(10), edgecolor="black")
plt.xlabel("Game Titles")
plt.ylabel("Global Sales (in millions)")
plt.title("Top 10 N64 Games by Global Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
# Using plt.bar to plot the global sales of the top 10 games specific to SNES Platform
plt.bar(snes['Name'].head(10), snes['Global_Sales'].head(10), color = "red", edgecolor= "black")
plt.xlabel("Game Titles")
plt.ylabel("Global Sales (in millions)")
plt.title("Top 10 SNES Games by Global Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
# Using plt.bar to plot the global sales of the top 10 games specific to NES Platform
plt.bar(nes['Name'].head(10), nes['Global_Sales'].head(10), color = "orange", edgecolor = "red")
plt.xlabel("Game Titles")
plt.ylabel("Global Sales (in millions)")
plt.title("Top 10 NES Games by Global Sales")
plt.xticks(rotation=90)
plt.show()
```
