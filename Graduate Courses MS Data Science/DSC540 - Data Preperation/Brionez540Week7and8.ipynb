**1. Transforming data using the to_numeric() function**

```python
# converting the age aolumn to numeric and forcing the values that cannot be converted to change to NaN values
df['Q3: AGE'] = pd.to_numeric(df['Q3: AGE'], errors = 'coerce')
```

```python
# checking that there are only numeric values remaining and NaN
df['Q3: AGE'].unique()
```

**2. Replacing missing / NaN values**

```python
# replacing NaN values with the median value of the column
df['Q3: AGE'] = df['Q3: AGE'].fillna(df['Q3: AGE'].median())
```

```python
# checking if there are any remaining NaN values in the 'Age' column
df['Q3: AGE'].isnull().values.any()
```

```python
# dropping NaN values from the 'Country' column 
df.dropna(subset = ['Q4: COUNTRY'], inplace = True)
```

```python
# importing additional tools 
from datetime import datetime
```

```python
# reviewing the dataframe
df2.head(1)
```

**1. Timestamp to period**

```python
# converting the 'Timestamp' column to the timestamp type
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'])
```

```python
# creating a new column in the dataframe that consists of the period from timestamp
df2['Period'] = df2['Timestamp'].dt.to_period('M')
```

```python
# using .pop() to remove the column from the end and save it to a variable
col = df2.pop('Period')
# re-insert the column to index 1
df2.insert(1, col.name, col)
```

```python
# reviewing the new dataframe
df2.head(1)
```

**2. Converting timestamp to string**

```python
# convert the 'timestamp' column to the timestamp data type
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'])
```

```python
# create a new column 'date' that is a string of the date potion of the timestamp column
df2['date'] = df2['Timestamp'].dt.strftime('%Y-%m-%d')
```

```python
# create a new column 'time' that is a string of the time portion of the timestamp column
df2['time'] = df2['Timestamp'].dt.strftime('%H:%M:%S')
```

```python
# checking the values in the new 'date' column
df2['date']
```

```python
# checking the values in the new 'time' column
df2['time']
```
